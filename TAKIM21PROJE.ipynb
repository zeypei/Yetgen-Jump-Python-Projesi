```python
from logging import exception
print("......Lezzet buradaya hosgeldiniz.....")

class kategoriler():
    
    malzemeler=[["kasar peyniri", "sucuk","domates sosu"],
              ["zeytin", "domates sosu", "biber", "mantar", "misir", "tofu"],
              ["kasar peyniri", "sucuk","domates sosu","biber", "mantar", "misir","sosis"],
              ["domates sosu","kasar","roka"]]

    pizzafiyat={"1-sucuklu pizza":40,
              "2-vegan":50,
              "3-bol malzemeli":45,
              "4-margarita":40}

    malzemelerh=[["ekmek", "cheedar peyniri", "domates", "kofte", "karamelize sogan", "mayonez","tursu","patates kizartmasi"],
                ["ekmek ","kofte","domates",  "karamelize sogan", "mayonez","tursu", "patates kizartmasi"],
                 ["ekmek","sinitzel","tursu","patates kizartmasi","cheedar"],
                 ["ekmek","falafel","marul","domates","yogurt sosu","patates kizartmasi"]]

    hamburgerfiyat={"1-chesseburger":50,
                  "2-kofteburger":60,
                  "3-tavukburger":55,
                   "4-veganburger":60}
    
    tatlifiyat ={"1-tiramisu":30, "2-trilece":40,"3- ekler":35, "4-dondurma":35}
   
    icecekfiyat = {'1-su':5, '2-kola':10 , '3-ice tea':10, '4-ayran':8, '5-gazoz':9, '6-milkshake':15, '7 -soda':7}
       
    sepet=[]

    @classmethod
    def pizza_islemleri(cls):
      
      for key,value in zip(cls.pizzafiyat,cls.malzemeler):
        print(f"{key:<15}",cls.pizzafiyat[key],"TL",value)
      
      pizza =  input("Hangi pizzayi istiyorsunuz?")

      if pizza not in list(cls.pizzafiyat.keys()):
        raise exception

      for a,b in zip(list(cls.pizzafiyat.keys()),cls.malzemeler):
        if pizza == a :
         secilen_pizza_malzemeleri = b.copy()
         print(a,b)

      while True:
       ekle_cikar = input("Ürün eklemek isterseniz '+' , ürün çıkarmak isterseniz '-' , Başka bir işlem yapmak istemiyorsanız '*' tuşlayınız")
       if ekle_cikar == "+":
        print("Ek ürünler : Sucuk , Biber , Mısır , Zeytin , Domatessosu , Kaşar , Sosis , Mantar")
        ek_urun_secenekleri = [ "sucuk" , "biber" , "mısır" , "zeytin" , "domatessosu" , "kaşar" , "sosis" , "mantar"]
        ek_urun =input("Eklemek istediğiniz ürünleri aralarında birer boşluk bırakarak giriniz:").lower()

        for i in ek_urun.split():
         if i not in ek_urun_secenekleri :
          raise exception

        secilen_pizza_malzemeleri.extend(ek_urun.split())
       elif ekle_cikar == "-":
         print(f"Sectiginiz pizzanin icindeki malzemeler {secilen_pizza_malzemeleri}")
         cikarilacak_urun=input("Cikarmak istediğiniz ürünleri aralarında birer boşluk bırakarak giriniz:").lower()

         for i in cikarilacak_urun.split():
           if i not in secilen_pizza_malzemeleri :
            raise exception

         for i in cikarilacak_urun.split():
           secilen_pizza_malzemeleri.remove(i)

       elif ekle_cikar == "*":
        print(f"Pizzanızdaki malzemeler : {secilen_pizza_malzemeleri}")
        cls.sepet.append(f"{pizza} , {cls.pizzafiyat[pizza]} , {secilen_pizza_malzemeleri}")
        print()
        print("Menu ekranina yonlendiriliyorsunuz") 
        break
      
       else:
         raise exception


    @classmethod
    def hamburger_islemleri(cls):
        
        for key,value in zip(cls.hamburgerfiyat,cls.malzemelerh):
         print(f"{key:<15}",cls.hamburgerfiyat[key],"TL",value)

        hamburger= input("Hangi hamburgeri istiyorsunuz?")

        if hamburger not in list(cls.hamburgerfiyat.keys()):
         raise exception

        for c,d in zip(list(cls.hamburgerfiyat.keys()), cls.malzemelerh):
          if hamburger == c:
            print(c,d)
            secilen_hamburger_malzemeleri=d.copy()

        while True:
         ekle_cikar = input("Ürün eklemek isterseniz '+' , ürün çıkarmak isterseniz '-' , Başka bir işlem yapmak istemiyorsanız '*' tuşlayınız")
         if ekle_cikar == "+":
          print("Ek ürünler : Biber ,Mantarsosu,  Domatessosu , Kaşar , Ekstraköfte","Falafel","Turşu")
          ek_urun_secenekleri = [ "biber" ,"mantarsosu",  "domatessosu" , "kaşar" , "ekstraköfte","falafel","turşu"]
          ek_urun = input("Eklemek istediğiniz ürünleri aralarında birer boşluk bırakarak giriniz:").lower()

          for i in ek_urun.split():
           if i not in ek_urun_secenekleri :
            raise exception


          secilen_hamburger_malzemeleri.extend(ek_urun.split())
         elif ekle_cikar == "-":
          print(f"Sectiginiz hamburgerin icindeki malzemeler {secilen_hamburger_malzemeleri}")
          cikarilacak_urun=input("Cikarmak istediğiniz ürünleri aralarında birer boşluk bırakarak giriniz:").lower()

          for i in cikarilacak_urun.split():
           if i not in secilen_hamburger_malzemeleri :
            raise exception

          for i in cikarilacak_urun.split():
           secilen_hamburger_malzemeleri.remove(i)

         elif ekle_cikar == "*":
          print(f"Hamburgerinizdeki malzemeler : {secilen_hamburger_malzemeleri} ")
          cls.sepet.append(f"{hamburger} , {cls.hamburgerfiyat[hamburger]} , {secilen_hamburger_malzemeleri}")
          print()
          print("Menu ekranina yonlendiriliyorsunuz") 
          break

         else:
          raise exception
      
    @classmethod
    def tatli_islemleri(cls):
      print(list(cls.tatlifiyat.keys()))
      tatli=input("Hangi tatliyi istersiniz?")
      if tatli not in list(cls.tatlifiyat.keys()):
        raise exception
      
      cls.sepet.append(f"{tatli} , {cls.tatlifiyat[tatli]}")


    @classmethod
    def icecek_islemleri(cls):
      print(list(cls.icecekfiyat.keys()))
      icecek= input("Hangi icecegi istersiniz?")
      if icecek not in list(cls.icecekfiyat.keys()):
        raise exception
      
      cls.sepet.append(f"{icecek} , {cls.icecekfiyat[icecek]}")


while True:
 try:
  islem = input("------MENÜ------ \n1.Pizza \n2.Hamburger \n3.Tatlı \n4.İçecekler \n---------------- \nAlmak istediğiniz ürünü seçiniz(Uygulamadan çıkmak için q'ya basınız):")
  if islem == "1":
    kategoriler.pizza_islemleri()
  elif islem == "2" :
    kategoriler.hamburger_islemleri()
  elif islem == "3":
    kategoriler.tatli_islemleri()
  elif islem == "4":
   kategoriler.icecek_islemleri()
  elif islem == "q":
   print("Sepetiniz : ")
   for i in kategoriler.sepet:
    print(i)
   print()
   print("Bizi tercih ettiğiniz için teşekkür ederiz :)")
   break
  else:
    raise exception

  print("-"*40)

 except:
   print("Hatalı giriş yaptınız. Lütfen tekrar deneyin.")
```
